```python
import matplotlib.pyplot as plt
import numpy as np

x=np.arange(0,10)
y=x**2
z=x**3

plt.plot(x,y)
plt.plot(x,z)
plt.xlabel('numbers')
plt.ylabel('squares and cubes')
plt.title('squares and cubes of numbers from 0 to 10')
```

```python
plt.scatter(x,y,marker='v',c='red')
plt.scatter(x,z,marker='o',c='black', alpha=0.5)
```

```python
plt.plot(x,y,'bo-')
```
